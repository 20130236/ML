#Task 1. 
For breast cancer dataset (https://tinyurl.com/3vme8hr3) which could be loaded from datasets in sklearn as follows:

```
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
cancer = datasets.load_breast_cancer()
```

*   1.1.	Apply SVM algorithm to above dataset using linear kernel.
*   1.2.	Compare the obtained results with other competitive algorithms (Logistic Regression, Decision Tree, kNN) based on metrics: accuracy, precision, recall, f1 measures.



````python
cancer = datasets.load_breast_cancer()
x = cancer['data']
y = cancer['target']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)

#SVM
SVM =  svm.SVC(kernel='linear')
SVM.fit(x_train, y_train) 
y_pred_svm = SVM.predict(x_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)
cm = confusion_matrix(y_test, y_pred_svm) 
print(cm)

# Decision Tree algorithm
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(x_train, y_train)
pred_labels_dtc = dtc.predict(x_test)
accuracy_dtc = accuracy_score(y_test, pred_labels_dtc)
precision_dtc = precision_score(y_test, pred_labels_dtc)
recall_dtc = recall_score(y_test, pred_labels_dtc)
f1_dtc = f1_score(y_test, pred_labels_dtc)


# kNN algorithm
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

# Logistic Regression algorithm
lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)


KNN = {'Accuracy:' : accuracy_knn, 'Precision:' :precision_knn, 'Recall:' : recall_knn, 'F1 score:' :f1_knn}
LR = {'Accuracy:' : accuracy_lr, 'Precision:' : precision_lr, 'Recall:' : recall_lr, 'F1 score:' : f1_lr}
DCT = {'Accuracy:' : accuracy_dtc, 'Precision:' : precision_dtc, 'Recall:' : recall_dtc, 'F1 score:' : f1_dtc}
SVM = {'Accuracy:' : accuracy_svm, 'Precision:' : precision_svm, 'Recall:' : recall_svm, 'F1 score:' : f1_svm}
knn2 = pd.Series(KNN)
lr2 = pd.Series(LR)
dct2 = pd.Series(DCT)
svm2 = pd.Series(SVM)

df = pd.DataFrame({'KNN': knn2,'LR': lr2,'DTC': dct2, 'SVM' : svm2})
df
````

#Task 2. 

*   1.1.	Perform SVM algorithm to **Iris dataset** using **linear kernel**.
*   1.2.	Compare the obtained results in 1.1 with SVM using other kernels (**Polynomial Kernel, Gaussian Kernel, Sigmoid Kernel, Radial Basis Function Kernel**). Some metrics could be used: accuracy, precision, recall, f1 measures





````python
data = datasets.load_iris()
````

````python
x = data['data']
y = data['target']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)

# linear
SVM_lr =  svm.SVC(kernel='linear')
SVM_lr.fit(x_train, y_train) 
y_pred_svm_lr = SVM_lr.predict(x_test)
accuracy_svm_lr = accuracy_score(y_test, y_pred_svm_lr)
precision_svm_lr = precision_score(y_test, y_pred_svm_lr, average = 'macro')
recall_svm_lr = recall_score(y_test, y_pred_svm_lr, average = 'macro')
f1_svm_lr = f1_score(y_test, y_pred_svm_lr, average = 'macro')

# poly
SVM_poly =  svm.SVC(kernel='poly',degree=10)
SVM_poly.fit(x_train, y_train) 
y_pred_svm_poly = SVM_poly.predict(x_test)
accuracy_svm_poly = accuracy_score(y_test, y_pred_svm_poly)
precision_svm_poly = precision_score(y_test, y_pred_svm_poly, average = 'macro')
recall_svm_poly = recall_score(y_test, y_pred_svm_poly, average = 'macro')
f1_svm_poly = f1_score(y_test, y_pred_svm_poly, average = 'macro')

# Gaussian
kernel = 1.0 * RBF(1.0)
gauss = GaussianProcessClassifier().fit(x_train,y_train)
y_pred_svm_gauss = gauss.predict(x_test)
accuracy_svm_gauss = accuracy_score(y_test, y_pred_svm_gauss)
precision_svm_gauss = precision_score(y_test, y_pred_svm_poly, average = 'macro')
recall_svm_gauss = recall_score(y_test, y_pred_svm_poly, average = 'macro')
f1_svm_gauss = f1_score(y_test, y_pred_svm_poly, average = 'macro')


# Sigmoid
SVM_sig =  svm.SVC(kernel='sigmoid')
SVM_sig.fit(x_train, y_train) 
y_pred_svm_sig = SVM_sig.predict(x_test)
accuracy_svm_sig = accuracy_score(y_test, y_pred_svm_sig)
precision_svm_sig = precision_score(y_test, y_pred_svm_sig, average = 'macro')
recall_svm_sig = recall_score(y_test, y_pred_svm_sig, average = 'macro')
f1_svm_sig = f1_score(y_test, y_pred_svm_sig, average = 'macro')

#Radial Basis Function 
SVM_rbf =  svm.SVC(kernel='rbf')
SVM_rbf.fit(x_train, y_train) 
y_pred_svm_rbf = SVM_rbf.predict(x_test)
accuracy_svm_rbf = accuracy_score(y_test, y_pred_svm_rbf)
precision_svm_rbf = precision_score(y_test, y_pred_svm_rbf, average = 'macro')
recall_svm_rbf = recall_score(y_test, y_pred_svm_rbf, average = 'macro')
f1_svm_rbf = f1_score(y_test, y_pred_svm_rbf, average = 'macro')

SVM_lr = {'Accuracy:' : accuracy_svm_lr, 'Precision:' : precision_svm_lr, 'Recall:' : recall_svm_lr, 'F1 score:' : f1_svm_lr}
SVM_poly = {'Accuracy:' : accuracy_svm_poly, 'Precision:' : precision_svm_poly , 'Recall:' : recall_svm_poly , 'F1 score:' : f1_svm_poly}
SVM_gauss = {'Accuracy:' : accuracy_svm_gauss, 'Precision:' : precision_svm_gauss, 'Recall:' : recall_svm_gauss, 'F1 score:' : f1_svm_gauss}
SVM_sig = {'Accuracy:' : accuracy_svm_sig, 'Precision:' : precision_svm_sig, 'Recall:' : recall_svm_sig, 'F1 score:' : f1_svm_sig}
SVM_rbf  = {'Accuracy:' : accuracy_svm_rbf, 'Precision:' : precision_svm_rbf, 'Recall:' : recall_svm_rbf, 'F1 score:' : f1_svm_rbf}

svm_lr = pd.Series(SVM_lr)
svm_poly =  pd.Series(SVM_poly)
svm_gauss =  pd.Series(SVM_gauss)
svm_sig =  pd.Series(SVM_sig)
svm_rbf  =  pd.Series(SVM_rbf)

df = pd.DataFrame({'linear kernel  ': svm_lr,'Polynomial Kernel  ': svm_poly,'Gaussian Kernel  ': svm_gauss, 'Sigmoid Kernel  ' : svm_sig, 'Radial Basis Function  ': svm_rbf})
style  = df.style.set_table_styles(
    [
      {"selector": "td, th", "props": [("border", "1px solid grey !important")]},
    ]
)

HTML(style.render())
````
